# Modelos com Espécies Binarizadas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from keras.models import Sequential
from keras.models import load_model
from keras.layers.normalization import BatchNormalization
from keras.layers.core import Dropout
from keras.layers.core import Dense
from keras.layers import Input
from keras.models import Model
from sklearn.preprocessing import MinMaxScaler
from keras.callbacks import EarlyStopping
import h5py
from keras.utils import plot_model
import shap
from sklearn.metrics import confusion_matrix, auc, roc_curve
import seaborn as sn
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

# Carregamento da Base

```python
X_train = pd.read_csv("Base_esp_onehot/X_train.csv",index_col="Unnamed: 0")
X_test = pd.read_csv("Base_esp_onehot/X_test.csv",index_col="Unnamed: 0")
y_train = np.load("Base_esp_onehot/y_train.npy",allow_pickle=True)
y_test = np.load("Base_esp_onehot/y_test.npy",allow_pickle=True)
```

```python
X_train.head()
```

## Abordagem A

```python
input_sape = X_train.shape[1]
batch = X_train.shape[0] // 10

inputs = Input(shape=(input_sape,))
x = BatchNormalization()(inputs)
x = Dropout(0.5)(x)

x = Dense(45, activation="elu")(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)

x = Dense(20, activation="elu")(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)

x = Dense(10, activation="elu")(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)

output = Dense(1, activation="sigmoid")(x)
modelO = Model(inputs,output)

modelO.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20,restore_best_weights=True)

history_modelO = modelO.fit(X_train, y_train,
                epochs=1000,
                batch_size=batch,
                shuffle=True,
                validation_split=0.3,
                callbacks=[es])
```

```python
plt.figure(figsize=(15,6))
plt.rc('font', size=14)
plt.plot(history_modelO.history['loss'],color='r',label="loss",linewidth=2)
plt.plot(history_modelO.history['val_loss'],label="val_loss",linewidth=2)
plt.xlabel(u"Épocas",size=16)
plt.ylabel("Binary Crossentropy",size=16)
plt.title("Modelo O: Erro durante o treinamento",size=18)
plt.legend()
plt.savefig('modelos_onehot/history_modelO.pdf', format="pdf", bbox_inches='tight')
plt.show()
```

```python
y_pred_O = modelO.predict(X_test)
y_pred_O = list(map(lambda p: 1 if p > 0.5 else 0, y_pred_O))

cm = pd.DataFrame(confusion_matrix(y_test,y_pred_O),index=['Nao_apareceu','Apareceu'],columns=['Nao_apareceu','Apareceu'])

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True,cmap='Blues')
plt.title("Matriz de confusão",size=20)
plt.xlabel("Classe prevista",size=18)
plt.ylabel("Classe real",size=18)
plt.yticks([0.3,1.4],['Não apareceu',"Apareceu"],size=14)
plt.xticks([0.5,1.5],['Não apareceu',"Apareceu"],size=14)
plt.savefig('modelos_onehot/cm_modelO.pdf', format="pdf", bbox_inches='tight')
plt.plot()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_O)
modelO_auc = auc(fpr, tpr)

plt.figure(figsize = (10,7))
plt.plot(fpr,tpr,label='Curva ROC: AUC = '+str(round(modelO_auc,2)),linewidth=2)
plt.plot([0,1],[0,1],'--k',label="Classificador aleatório")
plt.xlabel('False Positive Rate',size=16)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.ylabel('True Positive Rate',size=16)
plt.title("Modelo O")
plt.legend()
```

```python
modelO.summary()
```

```python
modelO.save("modelos_onehot/modelO.h5")
```

```python
X_train = pd.read_csv("Base_esp_onehot/X_train_window_3_mean.csv",index_col="Unnamed: 0")
X_test = pd.read_csv("Base_esp_onehot/X_test_window_3_mean.csv",index_col="Unnamed: 0")
y_train = np.load("Base_esp_onehot/y_train_window_3_mean.npy",allow_pickle=True)
y_test = np.load("Base_esp_onehot/y_test_window_3_mean.npy",allow_pickle=True)
```

```python
input_sape = X_train.shape[1]
batch = X_train.shape[0] // 10

p = 0.3

inputs = Input(shape=(input_sape,))
x = BatchNormalization()(inputs)
x = Dropout(p)(x)

x = Dense(45, activation="elu")(x)
x = BatchNormalization()(x)
x = Dropout(p)(x)

x = Dense(22, activation="elu")(x)
x = BatchNormalization()(x)
x = Dropout(p)(x)

output = Dense(1, activation="sigmoid")(x)
modelX = Model(inputs,output)

modelX.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20,restore_best_weights=True)

history_modelX = modelX.fit(X_train, y_train,
                epochs=1000,
                batch_size=batch,
                shuffle=True,
                validation_split=0.2,
                callbacks=[es])
```
